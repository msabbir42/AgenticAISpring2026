# Exercise 2

```python
import cv2
import os

cap = cv2.VideoCapture("/Users/wyd2hu/Library/CloudStorage/OneDrive-Personal/AI_Agent_Course/Topic 6/sample_video.mp4")
fps = cap.get(cv2.CAP_PROP_FPS)
interval = int(fps * 2)  # frame interval for ~2 seconds


frames = []
frame_num = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    if frame_num % interval == 0:
        frames.append(frame)
    frame_num += 1

cap.release()

for i, frame in enumerate(frames):
    cv2.imwrite(os.path.join("/Users/wyd2hu/Library/CloudStorage/OneDrive-Personal/AI_Agent_Course/Topic 6/image_frames", f"frame_{i:04d}.jpg"), frame)
```
